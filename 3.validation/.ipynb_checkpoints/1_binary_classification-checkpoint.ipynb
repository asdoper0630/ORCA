```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import pandas as pd
import tensorflow as tf
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE


import matplotlib.pyplot as plt
import collections
from sklearn.metrics import adjusted_mutual_info_score, \
normalized_mutual_info_score, mutual_info_score, adjusted_rand_score
import seaborn
import PIL

class_labels = {0 : 'high_risk', 1 : 'low_risk'}
```

```python
# set seed

seed=1055460630
np.random.seed(seed)
tf.set_random_seed(seed)
tf.compat.v1.set_random_seed(seed)

train_x = 'TCGA_trainingset.csv'

train_x = train_x.replace('\ufeff','')

# training / intracohort validation

# train the model

X,y = tf.contrib.learn.datasets.base.load_csv_with_header(
    filename=train_x,
    target_dtype=np.int,
    features_dtype=np.float32)

X_train,X_test,y_train,y_test = train_test_split(
    X,y,train_size = 0.8,test_size = 0.2,random_state = seed
)

test_set = X_test,y_test


X_test1,X_test2,y_test1,y_test2 = train_test_split(
    X_test,y_test,train_size = 0.5,test_size=0.5,random_state=seed
)

feature_columns = [tf.contrib.layers.real_valued_column("", dimension=22)]
config = tf.contrib.learn.RunConfig(tf_random_seed=1055460630)
classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[7,7],
                                            n_classes=2,
                                            config = config)

perm = []

for step in range(0,20):
    classifier.fit(x=X_train, y=y_train, steps=100)
    eval = classifier.evaluate(x=X_train,y=y_train)
    print (eval)
    current_loss = eval['loss']
    current_accu = eval['accuracy']

    y_pred = list(classifier.predict(X_test1, as_iterable=True))
    acc_count = 0
    for i,j in zip(y_pred,y_test1):
        if i == j:
            acc_count += 1
    acc_test1 = (acc_count/len(y_pred))
    print('accuracy : %.04f' % (100*(acc_count/len(y_pred)))+'%')

    y_pred = list(classifier.predict(X_test2, as_iterable=True))
    acc_count = 0
    for i,j in zip(y_pred,y_test2):
        if i == j:
            acc_count += 1
    print('accuracy : %.04f' % (100*(acc_count/len(y_pred)))+'%')
    acc_test2 = (acc_count/len(y_pred))
    perm.append((current_loss,current_accu,acc_test1,acc_test2))

# rand index
def rand_index(y_true,y_pred):
    n = len(y_true)
    a,b = 0,0
    for i in range(n):
        for j in range(i+1,n):
            if (y_true[i] == y_true[j]) & (y_pred[i] == y_pred[j]):
                a += 1
            elif (y_true[i] != y_true[j]) & (y_pred[i] != y_pred[j]):
                b += 1
            else: pass
    RI = (a+b) / (n*(n-1)/2)
    return RI

# apply new model x test1
y_pred = list(classifier.predict(X_test1, as_iterable=True))
y_pred1 = y_pred
print('rand index :',rand_index(y_test1,y_pred))
print('Adjusted rand score :',adjusted_rand_score(y_test1,y_pred))
print('Mutual info score :',mutual_info_score(y_test1,y_pred))
print('normalized_mutual_info_score :',normalized_mutual_info_score(y_test1,y_pred))
print('adjusted_mutual_info_score :',adjusted_mutual_info_score(y_test1,y_pred))

print('accuracy eval : %.04f' % (classifier.evaluate(x=X_test1,y=y_test1)["accuracy"]*100)+'%')

# apply new model x test2

y_pred = list(classifier.predict(X_test2, as_iterable=True))
y_pred2 = y_pred
print('rand index :',rand_index(y_test2,y_pred))
print('Adjusted rand score :',adjusted_rand_score(y_test2,y_pred))
print('Mutual info score :',mutual_info_score(y_test2,y_pred))
print('normalized_mutual_info_score :',normalized_mutual_info_score(y_test2,y_pred))
print('adjusted_mutual_info_score :',adjusted_mutual_info_score(y_test2,y_pred))

print('accuracy eval : %.04f' % (classifier.evaluate(x=X_test2,y=y_test2)["accuracy"]*100)+'%')

```

```python
# intercohort validation

test_x  = pd.read_csv('ORCA_testset.txt',sep='\t')

test_array  = []
for colname in test_x:
    test_array.append(list(test_x[colname]))
test_samples = np.array(test_array,dtype=float)

y = list(classifier.predict(test_samples, as_iterable=True))

result_df = pd.DataFrame({'Sample_ID':test_x.columns,'prediction_group':y})
result_df.to_csv('CIBERSORT_prediction.txt',sep='\t',index=False)


print(len(X_train),len(X_test1),len(X_test2))

seaborn.set(font_scale=1.3)

x_label         = []
loss_list       = []
intra_acc_list  = []
acc1_list       = []
acc2_list       = []

for num in range(0,20):
    curr_perm = perm[num]
    loss, intra_acc, acc1, acc2 = curr_perm
    x_label.append((num+1)*100)
    loss_list.append(loss)
    intra_acc_list.append(intra_acc)
    acc1_list.append(acc1)
    acc2_list.append(acc2)

plt.subplots()
plt.plot(x_label, loss_list)
plt.xlabel("steps")
plt.ylabel("Loss")
plt.savefig('./loss.tiff', dpi=300, bbox_inches='tight')

plt.subplots()
plt.plot(x_label, intra_acc_list, color="#EA0000")
plt.xlabel("steps")
plt.ylabel("Accuracy")
plt.savefig('./internal_accuracy.tiff', dpi=300, bbox_inches='tight')

plt.subplots()
plt.plot(x_label, acc1_list, color="orange")
plt.xlabel("steps")
plt.ylabel("Accuracy")
plt.savefig('./test1_accu.tiff', dpi=300, bbox_inches='tight')

plt.subplots()
plt.plot(x_label, acc2_list, color="#00CC00")
plt.xlabel("steps")
plt.ylabel("Accuracy")
plt.savefig('./test2_accu.tiff', dpi=300, bbox_inches='tight')

```
