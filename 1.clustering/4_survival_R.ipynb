```r
setwd('/root/cibersort/original/1.clustering')
f <- t(read.csv('clustered_sample.txt',sep='\t'))
library(survival)
f <- as.data.frame(f)
```

```r
# Hier_2
fit <- survfit(Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k2')
summary(fit)

plot(fit,ylab='Survival',xlab='Month',col=c('red','blue'),mark=3)
legend('topright',1,c("class1", "class2"), col = c("red", "blue"),lty=1:1)
a = survdiff(formula = Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k2')
a[['chisq']]
pchisq(a[['chisq']],df = 1, lower.tail=FALSE)
```

```r
# Hier_3 - significant. training required
fit <- survfit(Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k3')
summary(fit)
plot(fit,ylab='Survival',xlab='Month',col=c('red','blue','green'),mark=3)
legend('topright',1,c("class1", "class2","class3"), col = c("red", "blue","green"),lty=1:1)
a = survdiff(formula = Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k3')
a
pchisq(a[['chisq']],df = 2, lower.tail=FALSE)
```

```r

# Hier_4
fit <- survfit(Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k4')
summary(fit)
plot(fit,ylab='Survival',xlab='Month',col=c('red','blue','green','violet'),mark=3)
legend('topright',1,c("class1", "class2",'class3','class4'), col = c("red", "blue","green","violet"),lty=1:1)
a = survdiff(formula = Surv(f$'days',f$'vital_status',type='right')~f$'Hier_k4')
pchisq(a[['chisq']],df = 3, lower.tail=FALSE)
```
