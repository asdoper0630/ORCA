```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import PIL
sns.set(font_scale=1.3)

#seaborn Gr-Bk-Rd palette
pal = sns.diverging_palette(127.7, 12.2, l=53.4, n=255, center="dark")

pd.set_option('display.max_columns', 500)

from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test #logrank test를 적용함. multivariate_logrank_test, pairwise_logrank_test, survival_difference_at_fixed_point_in_time_test 이용 가능
import matplotlib.pyplot as plt
kmf = KaplanMeierFitter()

from scipy import stats


#data input
table = pd.read_csv("clustered_sample.txt",sep='\t').T


table['vital_status'] = (table['vital_status']==1.0)
table['days'] = abs((table['days']*30).astype(int))



# original survival

fig,ax = plt.subplots(figsize=(5,5))


# class 2 was low risk in visual.
# class 1,3 was merged with class 1

## 0 : high risk, 1 : low risk

print(list(table.Hier_k3))
table.Hier_k3 = abs(table['Hier_k3'].astype(int))
print(table.Hier_k3)


mask = (table['Hier_k3'] == 0)
kmf.fit(table['days'][mask],table['vital_status'][mask],label='cluster1')
kmf.plot(ax=ax, ci_show=False,color="#B55D60")

mask2 = (table['Hier_k3'] == 1)
kmf.fit(table['days'][mask2],table['vital_status'][mask2],label='cluster2')
kmf.plot(ax=ax, ci_show=False,color="#5975A4")

mask3 = (table['Hier_k3'] == 2)
kmf.fit(table['days'][mask3],table['vital_status'][mask3],label='cluster3')
kmf.plot(ax=ax, ci_show=False,color="#60B55D")

print(table['vital_status'][mask])

pvalue = 0.26592 # modified Log-Rank test - global p-value of (O-E)^2/E. see survival.R
ax.text(200,0.02,'P-value=%.5f'% pvalue)

ax.set_xlabel('Days')
ax.set_ylabel('Survival')

plt.savefig('./survival_hier3_each.tiff', dpi=300, bbox_inches='tight')

# binary survival

fig,ax = plt.subplots(figsize=(5,5))


# class 2 was low risk in visual.
# class 1,3 was merged with class 1

## 0 : high risk, 1 : low risk

print(list(table.Hier_k3))
table.Hier_k3 = abs(table['Hier_k3'].astype(int))
table.Hier_k3 = list(map(lambda x: x if x != 2 else 0, table.Hier_k3))
table.Hier_k3 = abs(table['Hier_k3'].astype(int))
print(table.Hier_k3)


mask = (table['Hier_k3'] == 0)
kmf.fit(table['days'][mask],table['vital_status'][mask],label='high_risk')
kmf.plot(ax=ax, ci_show=False,color="#B55D60")


mask2 = (table['Hier_k3'] == 1)
kmf.fit(table['days'][mask2],table['vital_status'][mask2],label='low_risk')
kmf.plot(ax=ax, ci_show=False,color="#5975A4")

result = logrank_test(table['vital_status'][mask], table['vital_status'][mask2], table['days'][mask], table['days'][mask2], alpha=.99)
pvalue = result.p_value
ax.text(200,0.05,'P-value=%.5f'% pvalue)


ax.set_xlabel('Days')
ax.set_ylabel('Survival')

plt.savefig('./survival_hier3_binary.tiff', dpi=300, bbox_inches='tight')
```
