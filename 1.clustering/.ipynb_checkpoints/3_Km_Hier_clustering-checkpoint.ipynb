```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sns.set(font_scale=1.3)

#seaborn Gr-Bk-Rd palette
pal = sns.diverging_palette(127.7, 12.2, l=53.4, n=255, center="dark")

import math
import numpy as np

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

def plotprocess():
    res['logPV'] = res['FDR'].apply(lambda x: -(math.log10(x)))

    res['type'] = 'etc'

    logfc = 1

    for i in res.index:
        if res.get_value(i, 'FDR') < 0.001:
            if res.get_value(i, 'logFC') > logfc:
                res.set_value(i, 'type', 'pl')
            elif res.get_value(i, 'logFC') < -logfc:
                res.set_value(i, 'type', 'mi')

# table list

with open('../HNSC_CIBERSORT.txt') as f:
    result = [f.readline().strip()]

# data input

df = pd.read_csv('../HNSC_CIBERSORT.txt', sep='\t', index_col=0)
df2 = df.loc['B cells naive':'Neutrophils']

df3 = df2.T

# Hierarchical AgglomerativeClustering
for num in range(2,5):

    x_std = StandardScaler().fit_transform(df2)
    features = x_std.T
    covariance_matrix = np.cov(features)
    cluster = AgglomerativeClustering(n_clusters=num, affinity='euclidean', linkage='ward')
    result.append('Hier_k'+str(num)+'\t'+'\t'.join(list(map(str,cluster.fit_predict(df3)))))

f1 = sns.clustermap(covariance_matrix, cmap=pal, method='ward').fig.suptitle('Hierarchical clustering', fontsize=24)
plt.savefig('./Hierarchical_clustering.png', dpi=300, bbox_inches='tight')



# K-means clustering
for num in range(2,4):
    kmeans = KMeans(n_clusters=num)
    kmeans = kmeans.fit(df3)
    labels = kmeans.predict(df3)
    kmeans.score(df3)
    result.append('Kmeans_k'+str(num)+'\t'+'\t'.join(list(map(str,labels))))
    # f1 = sns.clustermap(kmeans, cmap=pal, method='ward').fig.suptitle('Kmeans_clustering', fontsize=24)

Nc = range(1, 8)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(df3).score(df3) for i in range(len(kmeans))]
fig=plt.subplot()
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('K-means clustering')
plt.savefig('./K-means_clustering.tiff', dpi=300, bbox_inches='tight')


# add survival data

f = pd.read_csv('../HNSC_clinical.txt',sep='\t')
print(f)

sample = f['Sample ID']
f.index = sample
days_month = f['Overall Survival (Months)']
vital_status = f['Overall Survival Status']

with open('itemConsensus_BS_0104.txt') as f:
    lines = f.readlines()[1:]
CCdic = {}
for line in lines:
    lsp = line.strip().split('\t')
    CCdic[lsp[2]] = lsp[1]

CC = 'Consensus_k3'
status = 'vital_status'
days = 'days'

sample_index = result[0].strip().split('\t')

print(vital_status)
for samp in sample_index:
    CC += '\t'+ CCdic[samp]
    if vital_status[samp] == 'LIVING':
        status += '\t0'
    elif vital_status[samp] == 'DECEASED':
        status += '\t1'
    days += '\t'+str(days_month[samp])

result.append(CC)
result.append(status)
result.append(days)

# save
with open('clustered_sample.txt','w') as sv:
    sv.write('\n'.join(result))
```
